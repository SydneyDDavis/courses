## Pandas dataframes

### Creation

A DataFrame is a 2D table with labeled columns that can each hold different types of data. DataFrames are essentially a Python implementation of the types of tables you'd see in an Excel workbook or SQL database. DataFrames are the defacto standard data structure for working with tabular data in Python; we'll be using them a lot throughout the remainder of this guide.



```python
# create a dictionary with some different data types as values

my_dict = {"name" : ["Joe","Bob","Frans"],
           "age" : np.array([10,15,20]),
           "weight" : (75,123,239),
           "height" : pd.Series([4.5, 5, 6.1], 
                                index=["Joe","Bob","Frans"]),
           "siblings" : 1,
           "gender" : "M"}

df = pd.DataFrame(my_dict) # convert the dict to DataFrame

df
```

Also note that in the DataFrame above, the rows were automatically given indexes that align with the indexes of the series we passed in for the "height" column. If we did not use a series with index labels to create our DataFrame, it would be given numeric row index labels by default:



```python
# one without given index in height key : value
my_dict2 = {"name" : ["Joe","Bob","Frans"],
           "age" : np.array([10,15,20]),
           "weight" : (75,123,239),
           "height" :[4.5, 5, 6.1],
           "siblings" : 1,
           "gender" : "M"}

df2 = pd.DataFrame(my_dict2) # convert the dict to DataFrame

df2            
```

```python
# you can provide custom row labels when creating a
# DataFrame by adding the index argument:
df2 = pd.DataFrame(my_dict2, index = my_dict["name"] ) 
df2
```

This sort of logical True/False indexing is useful for subsetting data when combined with logical operations. For example, say we wanted to get a subset of our DataFrame with all persons who are over 12 years old. We can do it with boolean indexing:

```python
# create a boolean sequence with a logical comparison
boolean_index = df2["age"] > 12 # returns only rows w/ age over 12

df2[boolean_index] # should only return old bob and frans
```

```python
# you can also make it a ton easier by skipping the variable part...
df2[df2['age'] > 12] # same thing, no variable
```

```python
# let's do the same thing but filter for height equal to or over 5'
df2[df2['height'] >= 5.0] # returns same df as above
```

## Inserting missing data
You can insert missing values by simply assigning to containers. The actual missing value used will be chosen based on the dtype.

For example, numeric containers will always use NaN regardless of the missing value type chosen:

```python
# constructing series
s = pd.Series([1, 2, 3])
```

```python
# filling in with None
s.loc[0] = None
```

```python
# check this 
s
```

Likewise, datetime containers will always use NaT.

For object containers, pandas will use the value given:

```python
# constructing series
s = pd.Series(["a", "b", "c"])
```

```python
# filling index 0 with None
s.loc[0] = None
```

```python
# filling index 1 with NaN
s.loc[1] = np.nan
```

```python
# beautiful..?
s
```

### Calculations with missing data
Missing values propagate naturally through arithmetic operations between pandas objects. 

When summing data, NA (missing) values will be treated as zero.

If the data are all NA, the result will be 0.

Cumulative methods like cumsum() and cumprod() ignore NA values by default, but preserve them in the resulting arrays. To override this behaviour and include NA values, use skipna=False.

```python
df2
```

```python
df2["one"].sum()
```

```python
df2.mean(1)
```

```python
df2.cumsum()
```

```python
df2.cumsum(skipna=False)
```

```python
# the sum of an empty or all-NA Series or column of a DataFrame is 0.
pd.Series([np.nan]).sum()
```

```python
# here too
pd.Series([], dtype="float64").sum()
```

The product of an empty or all-NA Series or column of a DataFrame is 1.

```python
# let's demonstrate 
pd.Series([np.nan]).prod()
```

```python
# let's demonstrate here too 
pd.Series([], dtype="float64").prod()
```

### Replacing generic values
Often times we want to replace arbitrary values with other values.

replace() in Series and replace() in DataFrame provides an efficient yet flexible way to perform such replacements.

For a Series, you can replace a single value or a list of values by another value:

```python
# series!
ser = pd.Series([0.0, 1.0, 2.0, 3.0, 4.0])
```

```python
# replacing index 0 with value 5
ser.replace(0, 5)
```

```python
# same thing here as above just with more indeces
ser.replace([0, 1, 2, 3, 4], [4, 3, 2, 1, 0])
```

```python
# also here by mapping a dict
ser.replace({0: 10, 1: 100})
```

```python
# creating df3 
df3 = pd.DataFrame({"a": [0, 1, 2, 3, 4], "b": [5, 6, 7, 8, 9]})
```

```python
# replacing both col a, b index 0, 5 with '100'
df3.replace({"a": 0, "b": 5}, 100)
```

### String/regular expression replacement¶
Replace the ‘.’ with NaN (str -> str):



```python
# dict d
d = {"a": list(range(4)), "b": list("ab.."), "c": ["a", "b", np.nan, "d"]}
```

```python
# converting to dataframe
df = pd.DataFrame(d)
```

```python
df
```

```python
# replacing '.' with NaN
df.replace(".", np.nan)
```

```python
# now do it with a regular expression that removes surrounding whitespace (regex -> regex):
df.replace(r"\s*\.\s*", np.nan, regex=True)
```

```python
# replace a few different values (list -> list):
df.replace(["a", "."], ["b", np.nan])
```

```python
# list of regex -> list of regex:
df.replace([r"\.", r"(a)"], ["dot", r"\1stuff"], regex=True)
```

```python
# only search in column 'b' (dict -> dict):
df.replace({"b": "."}, {"b": np.nan})
```

```python
# same as the previous example, but use a regular expression for searching instead (dict of regex -> dict):
df.replace({"b": r"\s*\.\s*"}, {"b": np.nan}, regex=True)
```

```python
# you can pass nested dictionaries of regular expressions that use regex=True:
df.replace({"b": {"b": r""}}, regex=True)
```

```python
# alternatively, you can pass the nested dictionary like so:
df.replace(regex={"b": {r"\s*\.\s*": np.nan}})
```

```python
# You can also use the group of a regular expression match when replacing (dict of regex -> dict of regex)
# this works for lists as well.
df.replace({"b": r"\s*(\.)\s*"}, {"b": r"\1ty"}, regex=True)
```

```python
# you can pass a list of regular expressions, of which those that match will be replaced with a scalar 
# (list of regex -> regex).
df.replace([r"\s*\.\s*", r"a|b"], np.nan, regex=True)
```

All of the regular expression examples can also be passed with the to_replace argument as the regex argument. In this case the value argument must be passed explicitly by name or regex must be a nested dictionary. The previous example, in this case, would then be:



```python
df.replace(regex=[r"\s*\.\s*", r"a|b"], value=np.nan)
```
